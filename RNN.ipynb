```python
import pandas as pd
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import LabelEncoder
import numpy as np
```

```python
df = pd.read_csv("movies.csv");
df = df.dropna()
df.head()
```

```python
scaler = MinMaxScaler(feature_range=(0, 1))
df['Normalized_Score'] = scaler.fit_transform(df['score'].values.reshape(-1, 1))
df['Normalized_Votes'] = scaler.fit_transform(df['votes'].values.reshape(-1, 1))
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(2,)),
    tf.keras.layers.Dense(50, activation='relu'),
    tf.keras.layers.Dense(1)  
])
```

```python
model.compile(optimizer='adam', loss='mse')
```

```python
X = df[['Normalized_Score', 'Normalized_Votes']].values
y = df['Normalized_Score'].values
```

```python
model.fit(X, y, epochs=100, verbose=1)
```

```python
predictions = model.predict(X)
df['Prediction'] = predictions
threshold = 0.7
df['Recommended'] = df['Prediction'].apply(lambda x: 'Yes' if x >= threshold else 'No')
```
